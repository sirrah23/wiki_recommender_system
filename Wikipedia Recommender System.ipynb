# Building A Recommender System for Wikipedia

Here we are going to be building a recommender system for Wikipedia articles. We want to recommend articles to the Wikipedia users that edit articles. Our data is coming from the SNAP dataset on Wikipedia metadata. I have already filtered the data and seperated it into two files. One file contains information on what user edited which article, and the other file contains information about the articles themselves.

## Content-Based Recommendation

Lets start by reading in the two files we have and making them into dataframes

```python
import pandas as pd

anames = ['article_id','article_title','categories']
article_df = pd.read_table('/home/harris/article_info.main',sep=';',header=None,names=anames).drop_duplicates()
article_df = article_df.sort('categories',ascending=False).groupby('article_id',as_index=False).first()


unames = ['user_id','user_name','article_id']
user_df = pd.read_table('/home/harris/user_edits.main',sep=';',header=None,names=unames).drop_duplicates()
```

Lets grab just the user_id and article_id columns from this table and group that by user_id. That will give us a dictionary where each user is associated with the articles they have edited.

```python
user_groups = edit_df[[('user_id'),('article_id')]].groupby('user_id').groups
edit_df.ix[user_groups['75338'][1]]['categories'].split('|')
```

Now that we have this we can easily create a user profile for each user. We simply go through ever user in the user_groups dictionary and for each article they edited compute the mean occurance of every category.

```python
from collections import defaultdict

#function that creates the profile
def profile_creator(groups):
    user_profile = {}
    for name in groups:
        user_profile[name] = defaultdict(int)
        edits = groups[name]
        num_edits = float(len(edits))
        for edit in edits:
            #grab unique categories associated with each edit. We use set() to delete duplicates.
            categories = list(set(edit_df.ix[edit]['categories'].split('|')))
            for category in categories:
                user_profile[name][category] += 1
        for feature in user_profile[name]:
            user_profile[name][feature] /= num_edits
    return user_profile

#our dictionary of user profiles
user_profile = profile_creator(user_groups)
```

So now that we have the user profiles how do we recommend stuff to users? First we choose a distance measure. In our case it will be cosine distance. After that, we simply apply cosine distance to compare the distance between each user's profile and the available articles. After that we recommend to them the closest article. First lets define the cosine distance function.

```python
from math import sqrt

#Our vectors will technically be defaultdicts so we need to grab everything in both dictionaries.
#These two vectors are v1: the user profile and v2: the article profile
def cosine_dist(v1,v2):
    #First compute the cross product. We only need to check categories both of them have in common.
    cross_prod = 0
    for key in v1:
        cross_prod += v1[key] * v2[key]
    #Now compute the length of both vectors
    length_v1 = sqrt(len(v1))
    length_v2 = sqrt(len(v2))
    #Then the cosine similarity is
    cos_sim = cross_prod/(length_v1*length_v2)
    return cos_sim    
```

Now that we have a distance measure lets recommend an article to user '75338'. We technically have the table with all the information for the articles, we just need to turn that into "vectors" as well. When I say vector I mean a default dict with the corresponding values. We should ignore the articles that user '75338' has already edited in trying to recommend them another article so lets find those articles first.

```python
#Get the column of articles
this_users_articles = edit_df[edit_df['user_id'] == '75338']['article_id']
article_list = []
#Put every article_id into the a list
for i in range(len(this_users_articles)):
    article_list.append(this_users_articles.iloc[i])
#The list of user 75338's edited articles
article_list    
```

Now lets recommend user 75338 an article to edit! We are going to go through the article_df table and for each article we are going to create a defaultdict for that article's categories. After that we will compute the distance between user 75338's profile and the item profile. We will recommend to him the most similar article.

```python
#Give this function a user_profile and the list of articles they have already edited and it will recommend an article
def recommend(user_vec,article_list):
    #Number of rows in article_df table
    rows_article_df = len(article_df)
    #At the start we don't have a maximum distance so it is infinity
    max_sim = 0
    #We haven't recommended anything yet so keep this at -1
    recommended_article = -1
    #Loop through article_df table
    for i in range(rows_article_df):
        #Grab the ith row of the article_df table
        curr_row = article_df.iloc[i]
        #If user has already edited the article then skip it
        if curr_row['article_id'] in article_list:
            continue
        #Convert the categories from string to unique list to defaultdict    
        curr_categories = curr_row['categories']
        curr_categories = list(set(curr_categories.split('|')))
        curr_categories_vec = defaultdict(int)
        for item in curr_categories:
            curr_categories_vec[item] += 1.0  
        #Compute the distance between article_profile and user_profile   
        distance = cosine_dist(user_vec,curr_categories_vec)
        #Keep track of minimum distance and corresponding article
        if distance > max_sim:
            max_sim = distance
            recommended_article = curr_row['article_id']
    #Return article_id of recommended article        
    return recommended_article         
```

```python
#Lets now finally recommend an article to user 75338
curr_user_vec = user_profile['75338']
recommendation = recommend(curr_user_vec,article_list)
recommended = article_df[article_df["article_id"] == recommendation]
recommended
```

Now we will try to find a RMSE for this. First lets create a pivot table from the data we have where the users are the rows and the articles are the columns.

```python
pivoted_edit = edit_df.pivot_table(rows='user_id', cols='article_id',values = 'user_name', aggfunc=len, fill_value=0)
pivoted_edit
```

Now that we have the pivot table we want to split it so that the first 74 articles are in the training set and the last 19 articles are the test set. We have a 80/20 split.

```python
import random

#Split articles into testing and training sets after shuffling

all_articles = pivoted_edit.columns.values
random.shuffle(all_articles)
articles_train = all_articles[:56]
articles_test = all_articles[56:]  
```

Now we have the split we can simply generate user profiles again but this time use only the training set

```python
#function that creates the profile
def train_profile_creator(groups):
    user_profile = {}
    for name in groups:
        user_profile[name] = defaultdict(int)
        edits = groups[name]
        num_edits = float(len(edits))
        for edit in edits:
            #grab unique categories associated with each edit. We use set() to delete duplicates.
            article_id = edit_df.ix[edit]['article_id']
            if article_id in articles_train:
                categories = list(set(edit_df.ix[edit]['categories'].split('|')))
                for category in categories:
                    user_profile[name][category] += 1
            else:
                num_edits -= 1
        for feature in user_profile[name]:
            user_profile[name][feature] /= num_edits
    return user_profile

#our dictionary of user profiles
user_train_profile = train_profile_creator(user_groups)
```

```python
def sparsity_checker():
    for user in user_train_profile:
        row = pivoted_edit.ix[user]
        test_set = row.tail(37)
        for j in range(len(test_set)):
            if test_set.iloc[j] == 1:
                return True
    return False  

truth = False

while truth == False:
    all_articles = pivoted_edit.columns.values
    random.shuffle(all_articles)
    articles_train = all_articles[:56]
    articles_test = all_articles[56:]
    truth = sparsity_checker()
    user_train_profile = train_profile_creator(user_groups)
```

Now that we have the user_profiles on the training set we can recommend based on the test set

```python
def recommend_test_runner(epsilon):    
    #Recommendations for each user from the test set
    test_recommendations = defaultdict(int)
    
    #Give this function a user_profile and the list of articles they have already edited and it will recommend an article
    def recommend_test(user,user_vec,article_list,epsilon):
        #Number of rows in article_df table
        rows_article_df = len(article_df)
        #Loop through article_df table
        test_recommendations[user] = defaultdict(int)
        for i in range(rows_article_df):
            #Grab the ith row of the article_df table
            curr_row = article_df.iloc[i]
            #If article not in test set skip it
            article_id = curr_row['article_id']
            if article_id in article_list:
                continue
            #Convert the categories from string to unique list to defaultdict    
            curr_categories = curr_row['categories']
            curr_categories = list(set(curr_categories.split('|')))
            curr_categories_vec = defaultdict(int)
            for item in curr_categories:
                curr_categories_vec[item] += 1.0  
            #Compute the distance between article_profile and user_profile
            if user_vec != {}:
                distance = cosine_dist(user_vec,curr_categories_vec)
                #Keep track of minimum distance and corresponding article
                if distance > epsilon:
                    test_recommendations[user][article_id] = 1
        return
    
    #Recommendations for each user
    i = 0
    for user in user_train_profile:
        if i == 3400:
            break
        if user_train_profile[user]:
            recommend_test(user,user_train_profile[user],articles_train,epsilon)  
        i += 1
    return test_recommendations    
```

```python
recs = recommend_test_runner(0)
```

Now we have all the recommendations for users based on their training profile. Lets use this to compute the RMSE.

```python
def rmse(articles_test_set,test_results):
    for user in test_results:
        #numerator and denominator for rmse
        numerator = 0.0
        total = 1.0
        #Test set was last 37 values of the shuffled article list
        true_values = pivoted_edit.ix[user].tail(37)
        article_num = len(true_values)
        for i in range(article_num):
            if true_values.iloc[i] == 1:
                numerator += (true_values.iloc[i] - test_results[user][articles_test_set[i]])**2
                total += 1
    return sqrt(numerator/total)
```
